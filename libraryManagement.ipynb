```python
class Library:

    def __init__(self):
        self.file_name = "books.txt"
        self.file = open(self.file_name, "a+")

    def __del__(self):
        self.file.close()

    def list_books(self):

        self.file.seek(0)

        ### Create a list of books by separating each line with a comma
        books = [book.split(',') for book in self.file.read().splitlines()]
        for title, author, release_year, num_pages in books:
            print(f"Title: {title}, Author: {author}, Release Year: {release_year}, Pages: {num_pages}")

    def add_book(self):

        ### Get book information from user
        title = input("Enter book title: ")
        author = input("Enter author: ")
        release_year = int(input("Enter release year: "))
        num_pages = int(input("Enter number of pages: "))

        ### Add book information to file
        book_info = f"{title},{author},{release_year},{num_pages}\n"
        self.file.write(book_info)

        print(f"Book '{title}'.")

    def remove_book(self):

        ### Get the name of the book to delete
        title = input("Enter the title of the book to remove: ")
        self.file.seek(0)

        books = [book for book in self.file.readlines() if title not in book]

        self.file.seek(0)
        self.file.truncate()
        self.file.writelines(books)

        print(f"Book '{title}' removed.")



lib = Library() # Create an object

while True:     # Create a MENU

    print("*** MENU ***")
    print("1) List Books")
    print("2) Add Book")
    print("3) Remove Book")
    print("E) Exit")

    choice = input("Enter your choice (1/2/3/E): ")

    if choice == "1":
        lib.list_books()
    elif choice == "2":
        lib.add_book()
    elif choice == "3":
        lib.remove_book()
    elif choice == "E":
        print("Exit")
        break
    else:
        print("Invalid choice. Please enter 1 , 2 , 3 or E....")

```
